```python
import networkx as nx
from networkx.algorithms.centrality import degree_centrality,eigenvector_centrality,closeness_centrality
from networkx.algorithms.centrality import betweenness_centrality,subgraph_centrality

G = nx.read_weighted_edgelist('data/edgelist_dist10000')
```

```python
degree_df = pd.DataFrame(G.degree, columns = ['CODGEO', 'degree'])
degree_df['degree_centrality'] = degree_centrality(G).values()
degree_df['degree_centrality'] /= degree_df['degree_centrality'].max()

#degree_df['eigenvector_centrality'] = eigenvector_centrality(G).values()
#degree_df['eigenvector_centrality'] /= degree_df['eigenvector_centrality'].max()

degree_df['pagerank'] = nx.pagerank(G, alpha=0.9).values()
degree_df['pagerank'] /= degree_df['pagerank'].max()
```

```python
df_train['CODGEO'] = df_train['CODGEO'].astype(str)
```

```python
df_train = df_train.merge(degree_df, on = 'CODGEO')
```

```python
test_size = 0.5
indices = df_train.groupby('reg_id').indices

train_nodes = []
test_nodes = []

for k,v in indices.items() : 
    test_size_abs = int(len(v) * test_size) 
    random_choice = np.random.choice(len(v), len(v), replace=False)
    test_indices = v[random_choice[:test_size_abs]]
    train_indices =  v[random_choice[test_size_abs:]]
    
    train_nodes.extend(train_indices.tolist())
    test_nodes.extend(test_indices.tolist())

print(f'Training nodes : {len(train_nodes)}')
print(f'Testing nodes : {len(test_nodes)}')
```

```python
from tqdm import tqdm

def get_neigh_features(G, df_train, train_nodes = None, test_nodes = None) : 
    data = []

    if test_nodes is not None: 
        codes = df_train.iloc[test_nodes]['CODGEO']
        graph = G
        valid_neighbors = df_train.iloc[train_nodes]['CODGEO'].values.tolist()

    else : 
        codes = df_train.iloc[train_nodes]['CODGEO']
        graph = G.subgraph(codes)
        
    for code in tqdm(codes) : 
        neighbors = list(nx.neighbors(graph,code))
        if test_nodes is not None : 
            neighbors = list(set(neighbors).intersection(valid_neighbors))
        if len(neighbors) == 0 : 
            mean_votes = np.median(df_train.iloc[train_nodes]['y'])
            std_votes = np.std(df_train.iloc[train_nodes]['y'])
        else : 
            votes = df_train[df_train['CODGEO'].isin(neighbors)]
            mean_votes = np.mean(votes['y'])
            std_votes = np.std(votes['y'])


        data.append({'CODGEO' : code, "mean_votes" : mean_votes, "std_votes": std_votes})

    return pd.DataFrame(data)
```

```python
neigh_votes = get_neigh_features(G, df_train, train_nodes)
```

```python
test_neigh_votes = get_neigh_features(G, df_train, train_nodes = train_nodes, test_nodes = test_nodes)
```

```python
merged_train = df_train.merge(neigh_votes, on = 'CODGEO')
merged_test = df_train.merge(test_neigh_votes, on = 'CODGEO')

features = [i for i in merged_train.columns if i not in  ['y',"CODGEO", "reg_id", "dep_id","Votants_t2"]]

X_train = merged_train.reset_index(drop=True)[features]
y_train = merged_train['y'].reset_index(drop=True)
X_test = merged_test.reset_index(drop=True)[features]
y_test = merged_test['y'].reset_index(drop=True)
```
